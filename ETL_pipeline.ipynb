```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import sqlalchemy
import mysql.connector
import os
import boto3
from botocore.exceptions import ClientError
from boto3.session import Session
import psycopg2
from sqlalchemy import UniqueConstraint, text
```

```python
class news():
    
    def amazon_feed(self):
        url = 'https://docs.aws.amazon.com/marketplace/latest/userguide/document-history.rss'
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, features='xml')
        items = soup.findAll('item')
        news_items = []
        for item in items:
            new_item = {}
            new_item['Title'] = item.title.text
            new_item['Link'] = item.link.text
            new_item['Description'] = item.description.text
            new_item['PubDate'] = item.pubDate.text
    #         new_item['Creator'] = item.creator.text
            news_items.append(new_item)
        return news_items

    def world_feed(self):
        url = 'http://rss.cnn.com/rss/cnn_world.rss'
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, features='xml')
        items = soup.findAll('item')
        news_items = []
        for item in items:
            new_item = {}
            new_item['Title'] = item.title.text
            new_item['Link'] = item.link.text
            new_item['Description'] = item.description
            new_item['PubDate'] = item.pubDate
    #         new_item['Creator'] = item.creator.text
            news_items.append(new_item)
        return news_items
    
    def nytimes(self):
        url = 'https://rss.nytimes.com/services/xml/rss/nyt/World.xml'
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, features='xml')
        items = soup.findAll('item')
        news_items = []
        for item in items:
            new_item = {}
            new_item['Title'] = item.title.text
            new_item['Link'] = item.link.text
            new_item['Description'] = item.description.text
            new_item['PubDate'] = item.pubDate.text
    #         new_item['Creator'] = item.creator.text
            news_items.append(new_item)
        return news_items
    
    def yahoo(self):
        url = 'https://news.yahoo.com/rss/'
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, features='xml')
        items = soup.findAll('item')
        news_items = []
        for item in items:
            new_item = {}
            new_item['Title'] = item.title.text
            new_item['Link'] = item.link.text
#             new_item['guid'] = item.guid.text
            new_item['PubDate'] = item.pubDate.text
    #         new_item['Creator'] = item.creator.text
            news_items.append(new_item)
        return news_items
    
    def bbc_feed(self):
        url = 'http://feeds.bbci.co.uk/news/world/rss.xml#'
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, features='xml')
        items = soup.findAll('item')
        news_items = []
        for item in items:
            new_item = {}
            new_item['Title'] = item.title.text
            new_item['Link'] = item.link.text
            new_item['Description'] = item.description.text
            new_item['PubDate'] = item.pubDate.text
    #         new_item['Creator'] = item.creator.text
            news_items.append(new_item)
        return news_items
        
```
