#### Getting CRN numbers

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html, "html.parser")
```

```python
links=soup.find_all('a', {'href':True})
```

```python
len(links)
```

```python
links_selected=soup.find_all('a',{'href':True, 'data-toggle':'modal'})
```

```python
crn_links=[a_link for a_link in links_selected if a_link.text.strip().isnumeric()]
```

```python
crn_links=[]

for a_link in links_selected:
    clean_link_text=a_link.text.strip()
    print(clean_link_text)
    print(clean_link_text.isnumeric())
    if clean_link_text.isnumeric()==True: 
        crn_links.append(a_link)
    print("***")
```

```python
crn_links[0].text
```

```python
browser
```

```python
first_crn=browser.find_element_by_partial_link_text(crn_links[0].text.strip())
```

```python
first_crn
```

```python
def process_crn(crn_link):
    my_crn=browser.find_element_by_partial_link_text(crn_link.text.strip())
    my_crn.click()
    pop_content=browser.find_element_by_id('modalBodyContentDiv')

    df=pd.read_html(pop_content.get_attribute('innerHTML'))[0].T

    header=df.iloc[0]

    df=df[1:]
    df.columns=header
    
    pop_up_header=browser.find_element_by_class_name('modal-header')
    
    close_button=pop_up_header.find_element_by_tag_name('button')
    time.sleep(1)
    close_button.click()
    return df
```

```python
process_crn(crn_links[2])
```

```python
for link in crn_links:
    process_crn(link)
```
